```python
## export to folder

import cv2
import os
from datetime import datetime

# Path to the original video
video_path = "./test/test.mp4"

# Create timestamped output folder inside same folder as MP4
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
output_folder = os.path.join(os.path.dirname(video_path), timestamp)
os.makedirs(output_folder, exist_ok=True)

# Open video
cap = cv2.VideoCapture(video_path)

# Detect FPS of original video
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"Detected FPS: {fps}")

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break  # End of video

    # Save frame as zero-padded 5-digit JPG
    frame_filename = os.path.join(output_folder, f"{frame_count+1:05d}.jpg")
    cv2.imwrite(frame_filename, frame)

    frame_count += 1

cap.release()

print(f"✅ Done! {frame_count} frames saved in: {output_folder}")
print(f"ℹ️ FPS of original video: {fps}")
```

```python
## repack chuchu

import cv2
import os
from natsort import natsorted  # pip install natsort

# Folder containing frames (change this to the folder you want to repack)
frames_folder = "./runs/predict-seg/exp24"

# Output MP4 file path
output_video_path = os.path.join(frames_folder, "repacked_video.mp4")

# Get list of frames (sorted in natural order)
frames_list = natsorted([f for f in os.listdir(frames_folder) if f.lower().endswith(".jpg")])

if not frames_list:
    raise ValueError("❌ No JPG frames found in the folder.")

# Read first frame to get dimensions
first_frame = cv2.imread(os.path.join(frames_folder, frames_list[0]))
height, width, layers = first_frame.shape

# FPS (you can set manually or from previous FPS detection)
fps = 30.0  # Example, replace with detected FPS if known
print(f"Using FPS: {fps}")

# Define video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Write all frames
for frame_name in frames_list:
    frame_path = os.path.join(frames_folder, frame_name)
    img = cv2.imread(frame_path)
    video_writer.write(img)

video_writer.release()
print(f"✅ Video repacked and saved to: {output_video_path}")
```

```python
import os
import shutil

# Ask user for folder path
folder_path = input("Enter the folder path to compress: ").strip()

# Ensure the folder exists
if not os.path.isdir(folder_path):
    print("Error: The specified folder does not exist.")
else:
    # Get the folder name and define the zip file path
    folder_name = os.path.basename(folder_path)
    zip_file_path = os.path.join(os.getcwd(), f"{folder_name}.zip")

    # Compress the folder into a zip file
    shutil.make_archive(folder_name, 'zip', folder_path)

    print(f"Folder compressed successfully! Saved as: {zip_file_path}")
```
